# Investigating the Iris Dataset #

## Introduction ##

The iris dataset is a model 'toy' dataset within the scikit-learn website, which provides data about the size of three different species of iris. The species of interest include Setosa, Versicolor and Virginica. A total of 150 plants were measured in this study, with an even spread of 50 plants from each of the three species type. Four key measurements were taken from each plant, including the petal length, petal width, sepal length and sepal width. 
The following data exploration was conducted to identify whether these different measurements could be used to predict the iris species. This may offer a more facile method of identification of iris species in the wild, by performing a brief assessment by measuring petal length or width to determine the identity of an unknown iris plant of interest.

## Methods and Results ##

### 1. Loading the iris dataset ###

To begin exploring the iris dataset, the data was first imported into JupyterLab from scikit-learn. Pandas was also imported to perform some initial visualisation of the data in question.

### 7. Creating a Linear Model ###

A linear regression model was then created, using petal length to predict petal width, as these variables showed a strong positive correlation when observing the correlation analysis plots.

To generate the desired linear regression model, the relevant packages were installed from scikit-learn (sklearn) for generating such a model.

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept=True)
model
```

The petal length and petal width were then fitted to a linear regression model.

```python
import seaborn as sns

sns.relplot(data=iris_df, x="petal length (cm)", y="petal width (cm)")
```

```python
X = iris_df[["petal length (cm)"]]
y = iris_df["petal width (cm)"]
```

```python
model.fit(X, y)
```

Once the data was fitted to the model, the model was used to create predictions of new Y values using new hypothetical data between the minimum and maximum original datapoints of the X variables. The minimum and maximum X values (petal length) were first identified in order to set the correct parameters in the model predictions.

```python
iris_df["petal length (cm)"].min()
```

```python
iris_df["petal length (cm)"].max()
```

```python
pred = pd.DataFrame({"petal length (cm)": [1, 6.5]}) 
pred["petal width (cm)"] = model.predict(pred)
pred
```

The new data created along with the resulting predictions from the model was then plotted (blue) on a simple scatterplot to visualise the data points with respect to the model red dashed line.

```python
import seaborn as sns
sns.relplot(data=iris_df, x="petal length (cm)", y="petal width (cm)")
sns.lineplot(data=pred, x="petal length (cm)", y="petal width (cm)", c="red", linestyle=":")
```

The gradient and intercept of the modelled data was then extracted to generate the model's line of best fit equation:


```python
print(" Model gradient: ", model.coef_[0])
print("Model intercept:", model.intercept_)
```

Therefore the line equation was : y = 0.42x - 0.36

### 8. Testing the New Model ###

To determine the quality of the fit of the data to the linear regression model that was generated, the iris data was then split to create two data subsets: the training dataset, and the testing dataset.

```python
import pandas as pd

X = iris_df[["petal length (cm)"]]
y = iris_df["petal width (cm)"]
```

```python
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(X, y, random_state=42)
```

These train and test datasets were then plotted on a scatterplot to visualise the split in the original data.

```python
import seaborn as sns

iris_df.loc[train_X.index, "train/test"] = "train"
iris_df.loc[test_X.index, "train/test"] = "test"

sns.relplot(data=iris_df, x="petal length (cm)", y="petal width (cm)", hue="train/test")
```

The train data was then applied to the model to train the linear regression model. The resulting model score was then obtained to validate the model efficacy in explaining variation in the data when introduced to the test dataset.

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept=True)
model.fit(train_X, train_y)
```

```python
model.score(test_X, test_y)
```

The >0.9 model score indicated an effective model in preciting petal width according to petal length.

### 9. Nearest Neighbours ###

Another supervised regression technique was used to categorise the data, determining whether petal length and petal width could act as good predictors of the species class of each individual iris plant studied.
The iris data was first visualised on a scatter plot, separating each species type by colour.

```python
import seaborn as sns

sns.scatterplot(data=iris_df, x="petal length (cm)", y="petal width (cm)", hue="species", palette="Dark2")
```

```python
from pandas import DataFrame
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

X = iris_df
X = X[["petal length (cm)", "petal width (cm)"]]  
y = iris_df["species"]

train_X, test_X, train_y, test_y = train_test_split(X, y, random_state=42)
```

The relevant Nearest Neighbours model was initially imported, and the iris data was split into train and test data subsets which could be used to train the model, then test it in the model's ability to predict the iris species based upon the legnth and width of the petals. To assess the number of n neighbours that would more effectively split the data without overfitting the data, a range of n neighbour values between 1 and 40 were tested.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

parameters = {
    "n_neighbors" : range(1, 40),
}
clf = GridSearchCV(KNeighborsClassifier(), parameters).fit(train_X, train_y)
```

```python
cv_results = DataFrame(clf.cv_results_)
cv_results = cv_results.sort_values(["rank_test_score", "mean_test_score"])
cv_results.head()[["param_n_neighbors", "mean_test_score", "std_test_score", "rank_test_score"]]
```

The most successful number of neigbours tested were 7, 9, 10, 11 and 19, each sharing the same standard deviations and rank test scores, therefore any of these neigbour values could hav ebeen used for the same model efficacy.

```python
cv_results.plot.scatter("param_n_neighbors", "mean_test_score", yerr="std_test_score")
```


The predictions of the model were then plotted and the boundaries which were determined by the model between the petal sizes of different species were visualised.

```python
from sklearn.inspection import DecisionBoundaryDisplay
import seaborn as sns

DecisionBoundaryDisplay.from_estimator(clf, X, cmap="Pastel2")
sns.scatterplot(data=X, x="petal length (cm)", y="petal width (cm)", hue=y, palette="Dark2")
```

A very distint boundary was created between Setosa and Versicolor, as none of the data points overlapped each other's boundary, a strong boundary was also established between Virginica and Verisoclor, despite a few overlaps between datapoints, which avoided overfitting of the data to the model.

The model score was then extracted to assess the efficacy of the model in predicting iris species from their petal length and width.

```python
clf.score(test_X, test_y)
```

## Conclusion ##

A linear regression model was successfully generated with a model score of >0.9 to predict the petal width of iris plants according to input petal length. A nearest neighbours model was also generated to predict iris species based upon input petal size (petal length and width) with a model score of 1. Since the Setosa species had far smaller leaves than the Virginica and Versicolor species, but the Virginica and Versicolor species had different sized leaves which were sufficent to predict species type, yet these groups were less distinct, this may suggest that Virginica and Versicolor species may be more closely phylogenetically related compared with Setosa species. Further data analysis would need to be conducted in the future using more data beyond just petal sizes and sepal sizes; DNA barcode sequences, such as the ITS region would need to be used to generate phylogenetic trees to fully assess phylogneteic relationships between these different species. Nevertheless, it would be interesting to compare phylogenetic analyses to these models to determine whether phylogenetic relationship also relates to similar or different petal sizes.
